```python
y = dumy_y.values
y.shape
```

### Build the model

Standarize the data

```python
from keras import models
from keras import layers
from keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping
from keras import metrics
from sklearn.model_selection import train_test_split
```

```python
model = models.Sequential()
model.add(layers.Dense(32 , activation='relu' , input_dim=X_numpy.shape[1] ))
model.add(layers.Dense(16 , activation='relu' ))
model.add(layers.Dense(8 , activation='relu' ))
model.add(layers.Dense(y.shape[1], activation='sigmoid'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=[metrics.categorical_accuracy])
from tensorflow.keras.callbacks import EarlyStopping
monitors = EarlyStopping(min_delta=1e-3,patience=5 ,
                         verbose=1 ,restore_best_weights =True)
x_train,x_test , y_train , y_test =train_test_split(df4.values,y,random_state=1)

history=model.fit(x_train,y_train ,validation_data=(x_test,y_test),callbacks=[monitors],epochs=100)
```

```python
history_dict = history.history
history_dict.keys()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## <b> Loss 

```python
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values)+ 1)
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## <b> Accuracy 

```python
plt.clf()
acc = history.history['categorical_accuracy']
val_acc = history.history['val_categorical_accuracy']
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
